```python
year = ['2015', '2016', '2017', '2018', '2019', '2020', '2021']
dataset = {}
```

```python
for i in city:
    data = []
    for j in year:
        temp = df.loc[i,j]
        result = round(int(temp[1])/int(temp[0])*100)
        data.append(result)
        
    dataset[i] = data
```
