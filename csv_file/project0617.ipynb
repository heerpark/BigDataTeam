```python
def sortbaby(df, sigidx, intersection):
    
    #변수설정
    df_sig = list(df.iloc[:,sigidx])
    del_list = []
    column_name = df.columns[sigidx]
    #없애보자
    for i, sig in enumerate(df_sig):
        if sig in intersection:
            pass
        else:
            del_list.append(i)
    drop_data = df.drop(del_list, axis=0)
    drop_data = drop_data.drop_duplicates(subset=column_name)
    
    #drop_data를 sigidx기준으로 정렬
    sorted_df = drop_data.sort_values(by=column_name)
    
    return sorted_df
```
