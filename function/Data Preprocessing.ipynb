## PreProcessing함수 
### data와 data에서 sig코드가 있는 열의 idx를 입력하면 다음과 같은 동작을 한다.
#### 1. forium에 입력할 수 있도록 data를 전처리 해준다. 
#### 2. forium에서 표시가 안되는 지역을 리스트로 반환해준다.

```python

def PreProcessing(data, sigidx):
    del_list = []
    no_forium_idx = []
    no_forium = []
    data_sig = list(data.iloc[:,sigidx])
    
    #data의 sig코드가 json파일의 sig코드에 없으면 data에서 해당 행을 삭제
    for i,sig in enumerate(data_sig):
        if sig in json_sig:
            pass
        else:
            del_list.append(i)
    drop_data = data.drop(del_list, axis=0)
    drop_data_sig = list(drop_data.iloc[:,sigidx])
    
    
    #json파일의 sig코드중에 drop data에 없어 forium에 표시가 안되는 지역 추출
    for i,sig in enumerate(json_sig):
        if sig in drop_data_sig:
            pass
        else:
            no_forium_idx.append(i)
    for i in no_forium_idx:
        buf = str(json_name[i]) + ":" + str(json_sig[i])
        no_forium.append(buf)
    
    #forium에서 결측치 없이 표현 할 수 있도록 데이터의 평균 값으로 만들어 주고 
    
    
    return drop_data, no_forium
```

```python
def PreProcessing2(data, sigidx, dataidx):
    del_list = []
    no_forium_idx = []
    no_forium = []
    data_sig = list(data.iloc[:,sigidx])
    no_forium_sig = []
    data_mean_list = []
    
    #data의 sig코드가 json파일의 sig코드에 없으면 data에서 해당 행을 삭제
    for i,sig in enumerate(data_sig):
        if sig in json_sig:
            pass
        else:
            del_list.append(i)
    drop_data = data.drop(del_list, axis=0)
    drop_data_sig = list(drop_data.iloc[:,sigidx])
    
    
    #json파일의 sig코드중에 drop data에 없어 forium에 표시가 안되는 지역 추출
    for i,sig in enumerate(json_sig):
        if sig in drop_data_sig:
            pass
        else:
            no_forium_idx.append(i)
    for i in no_forium_idx:
        buf = str(json_name[i]) + ":" + str(json_sig[i])
        no_forium_sig.append(json_sig[i])
        no_forium.append(buf)
        
    
    #forium에서 결측치 없이 표현 할 수 있도록 dataframe 재구성
    data_mean = data.iloc[:,dataidx].mean()
    for_length = len(no_forium_sig)
    for i in range(for_length):
        data_mean_list.append(data_mean)
    sig_name = data.columns[sigidx]
    data_name = data.columns[dataidx]
    new_data = pd.DataFrame({sig_name:no_forium_sig, data_name:data_mean_list})
    last_data = pd.concat([drop_data, new_data], ignore_index=True)
    #last_data = drop_data.append(new_data, ignore_index=True)
    
    
    return drop_data, last_data, no_forium
```
